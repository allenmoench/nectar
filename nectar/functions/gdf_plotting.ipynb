```python
nm_download = gpd.read_file("https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_county_500k.zip")
nm_download
```

```python
nm_counties = nm_download[nm_download['STATEFP'] == '35']
nm_counties
```

```python
nm_counties.to_crs(26913) #use EPSG 26913 (UTM 13N - NAD83)
counties_plot = nm_counties.hvplot(geo=True, color='white', line_color='black', line_width=1, tiles='CartoLight', frame_width=500)
counties_plot
```
